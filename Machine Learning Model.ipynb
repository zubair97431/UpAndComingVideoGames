```python
import pandas as pd 
import numpy as np 
```

```python
# Import our files 
file_to_load = './vgsales.csv'
```

```python
# Load the csv 
video_game_sales= pd.read_csv(file_to_load)

video_game_sales
```

```python
# File to load 
file_to_load = './steam_games.csv'
```

```python
games_features = pd.read_csv(file_to_load)

games_features
```

```python
games_features = games_features.rename(columns={'name': 'Name'})
```

```python
games_features = games_features[['Name', 'all_reviews', 'popular_tags', 'game_details', 'genre']]
```

```python
games_features
```

```python
merged_df=pd.merge(video_game_sales,games_features, on='Name',how='outer')

```

```python
merged_df
```

```python
# We will begin cleaning the data by dropping null values, and converting certain columns into numberical if needed
```

```python
merged_df = merged_df.dropna(axis='columns', how='all')
```

```python
merged_df
```

```python
merged_df = merged_df.dropna()
```

```python
merged_df
```

```python
merged_df = merged_df.drop(['Publisher', 'genre', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'NA_Sales', 'Rank', 'Year', 'game_details'], axis=1)
```

```python
# Create our features
X = merged_df.drop(["popular_tags"], axis=1)

# Create our target
y = merged_df["popular_tags"]
```

```python
X = pd.get_dummies(X)
```

```python
X.describe()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
# Train the data
classifier.fit(X_train, y_train)
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```
